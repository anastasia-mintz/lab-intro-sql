```python
# How many distinct (different) actors' last names are there?


data = pd.read_sql_query('''SELECT COUNT(DISTINCT last_name) AS 'Different Actor Last Names' FROM sakila.actor''', engine)
data.head()
```

```python
# Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
data = pd.read_sql_query('''SELECT *, month(rental_date) AS 'Month',
CASE
WHEN weekday(rental_date) < 6 then 'Weekday'
ELSE 'Weekend'
END AS 'Weekday'
FROM sakila.rental
LIMIT 20''', engine)
data
```

```python
# Get all films with ARMAGEDDON in the title.

data = pd.read_sql_query('''SELECT title FROM sakila.film_text WHERE title LIKE '%ARMAGEDDON%' ''', engine)
data.head()
```

```python
# Get 10 the longest films.
data = pd.read_sql_query('SELECT title, length AS "Duration" FROM sakila.film ORDER BY length DESC', engine)
data.head(10)
```

```python
# How many films include Behind the Scenes content?
data = pd.read_sql_query('''SELECT title, special_features FROM sakila.film
WHERE special_features LIKE '%Behind the Scenes%' ''', engine)
data

#538 rows = 538 films
```

```python
# Which kind of movies (rating) have a mean duration of more than two hours?
data = pd.read_sql_query('''SELECT rating, AVG(length) AS 'Movie Duration' FROM sakila.film GROUP BY rating HAVING AVG(length) > 120''', engine)
data
```

```python
# Rank films by length (filter out the rows that have nulls or 0s in length column). 
#In your output, only select the columns title, length, and the rank.
data = pd.read_sql_query('''SELECT title, length,
CASE
WHEN length <= 100 THEN '1'
WHEN length <= 150 THEN '2'
ELSE '3'
END AS 'Rank'
FROM sakila.film
WHERE (length IS NOT NULL) AND (length <> ' ')''', engine)
data
```
